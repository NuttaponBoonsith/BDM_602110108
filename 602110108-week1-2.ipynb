```python
def multi(start,end):
    for i in range(start,end+1): # end+1 ให้มันจบที่ปลายจริงๆ
        for j in range(12): # range 0 ถึง 11 เลย+1ข้างล่างให้เป็น1ถึง12
            print(str(i)+' x '+str(j+1)+' = '+str(i*(j+1)))
        print('\n')
        
multi(2,4)
```
